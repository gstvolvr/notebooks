# Big O Complexities

We measure the perfomance of algorithms using their time and space complixities, which are represented using `Big O` notation. It usually reprents the worst case scenario of how long, or how much space and algorithm will take up depending on it's input

```python
import matplotlib.pyplot as plt
plt.style.use("seaborn")
import numpy as np

O_1 = lambda n: 1 
O_n = lambda n: n 
O_log_n = lambda n: np.log(n)
O_n_log_n = lambda n: n * np.log(n)
O_n_sq = lambda n: np.power(n, 2)
O_n_cb = lambda n: np.power(n, 3)
O_2_n = lambda n: np.power(2, n)
O_n_fc = lambda n: np.math.factorial(n)

funcs = [(O_1, "constant"), 
         (O_log_n, "log n"),  
         (O_n, "linear"), 
         (O_n_log_n, "n log n"), 
         (O_n_sq, "n ^ 2"), 
         (O_n_cb, "n ^ 3"), 
         (O_2_n, "2 ^ n"),
         (O_n_fc, "n!")]

fig = plt.figure(figsize=(18, 10))
r = np.arange(1, 6)

for f, l in funcs:
    plt.plot(r, list(map(f, r)), label=l)
plt.legend();
```
