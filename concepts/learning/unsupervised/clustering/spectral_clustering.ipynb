# Spectral Clustering

1. Represent graph as an `adjacency matrix` $A \in R^{m\times m}$
  * an `adjacency matrix` can be seen as a `node by node` sparse matrix where the values are `1` if there's a link.
2. Form a special matrix $L = D - A$, the graph `laplacian`
  * Depends on the `degree matrix` $D$, and the adjacency matrix $A$
3. Compute $k$ eigenvectors $v^1,v^2,\dots,v^k$ of $L$ corresponding to the $k$ <i>smallest</i> eigenvalues ($k<<m$)
  * the `eigen vectors` are the vectors such that the map of the `eigen vector` under the `L` matrix is going to be itself
4. run `k-means` algorithm on $Z = (v^1,v^2,\dots,v^k)$ by treating each point as a new data point. 


The can find the number of clusters by looking for all the `eigen vector's` whose `eigen value` is 0.

```python
from sklearn.cluster import SpectralClustering
from sklearn.datasets import make_circles
import seaborn as sns
sns.set()

X, y = make_circles(n_samples=250)
l = SpectralClustering(n_clusters=2, affinity="nearest_neighbors").fit_predict(X)

sns.scatterplot(X[:, 0], X[:, 1], hue=l);
```
