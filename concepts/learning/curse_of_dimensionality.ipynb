# Curse of Dimensionality
As the number of features or dimensions grows, the amount of data we need to generalize accurately, grows `exponentially`. This is due to an increase in sparsity, which happens because the volume of the space increases with dimensionality. 

```python
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
plt.style.use("seaborn")
```

### Visualizing the space increase

```python
fig, ax = plt.subplots(nrows=1, ncols=2, figsize=(18, 6))
X = np.random.randint(1, 10, size=(10, 1))
ax[0].scatter(X, np.ones(X.shape[0]))
ax[0].title.set_text("1) Say you start off with a single feature. \nIt can be shown on a sinlge number line.")

X = np.random.randint(1, 10, size=(10, 2))
ax[1].scatter(X[:, 0], X[:, 1])
ax[1].title.set_text("2) Then you add a new feature, and suddenly the 'space' grows.")

fig = plt.figure(figsize=(18, 6))
X = np.random.randint(1, 10, size=(10, 3))
ax = fig.add_subplot(111, projection='3d')
ax.scatter(X[:, 0], X[:, 1], X[:, 2])
ax.title.set_text("3) And so on...")
```
