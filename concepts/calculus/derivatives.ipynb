# Derivatives
It's a measure of how sensitive a function is to small changes to it's input. 

#### Product Rule
$$\frac{d}{dx}[f(x)g(x)] = f'(x)g(x) + f(x)g'(x)$$

### Approxmiating Derivatives

$$f'(a) = lim_{h\rightarrow 0} \frac{f(a+h) - f(a)}{h}$$

```python
from autograd import grad
import autograd.numpy as np

def derive(f, a):
    h = 1e-7 # some tiny value
    num = f(a + h) - f(a)
    return num / h

eps = 1e-1
for i in range(10, 20):
    f = lambda x: x ** 3
    assert abs(derive(f, i) - grad(f)(float(i))) < eps
    
    f = lambda x: x ** 5 + x ** 2 + 10
    assert abs(derive(f, i) - grad(f)(float(i))) < eps
    
    f = lambda x: 2 ** x
    assert abs(derive(f, i) - grad(f)(float(i))) < eps
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
sns.set()
r = range(-100, 100)
%matplotlib inline

def plot_functions(f, fp, fpp):
    plt.figure(figsize=(18, 6))
    if f:
        plt.plot(list(map(f, r)), '-.', label="function", alpha=1.);
    if fp:
        plt.plot(list(map(fp, r)), '--', label="1st derivative", alpha=0.5);
    if fpp:
        plt.plot(list(map(fpp, r)), '.', label="2nd derivative", alpha=0.2)
    plt.legend();
```

### Common derivatives

$$ \frac{d}{dx} x^2 = 2 x $$

```python
f = lambda x: x ** 2
fp = lambda x: 2 * x
fpp = lambda x: 2
plot_functions(f, fp, fpp)
```

---

$$ \frac{d}{dx} e^x = e^x$$

```python
f = lambda x: np.math.exp(x)
fp = lambda x: np.math.exp(x)
fpp = lambda x: np.math.exp(x)
plot_functions(f, fp, fpp)
```

---

$$ \frac{d}{dx} log(x) = \frac{1}{x}$$

```python
eps = 1e-4
f = lambda x: np.log(x + eps) if (x > 0) else None
fp = lambda x: 1 / (x + eps)
fpp = lambda x: -1 / (x ** 2 + eps)
plot_functions(f, fp, fpp)
```

---

$$ \frac{d}{dx} sin(x) = cos(x)$$


$$ \frac{d}{dx} cos(x) = -sin(x)$$

```python
f = lambda x: np.math.sin(x)
fp = lambda x: np.math.cos(x)
fpp = lambda x: -np.math.sin(x)

plot_functions(f, fp, fpp)
```

---

$$ \frac{d}{dx} 2^x = 2^x log(2)$$

```python
f = lambda x: 2 ** x
fp = lambda x: 2 ** x * np.log(2)

plot_functions(f, fp, None)
```
