# Beta Distribution
Continuos probability distribution defined on the interval `[0, 1]`. It's the `conjugate prior` to the [bernoulli](bernoulli_distribution) and [binomial](binomial_distribution) distributions. 

`conjugate prior` — if we have an unknown probability (e.g. the bias of a coin) that we're estimating, then the likelihood induced on the unknown bias follows a `beta distribution`. 

We use the following variables:
* $\alpha$ — shape parameter 
* $\beta$ — shape parameter

When we set $\alpha = 1$ and $\beta = 1$, the distribution will act as if we are flipping a coin. 

$$\text{Beta}(\mu|a, b) = \frac{\Gamma (a + b)}{\Gamma(a) + \Gamma(b)}\mu^{a-1}(1-\mu)^{b-1}$$

$$\mathbb{E}(\mu) = \frac{a}{a+b}$$

$$\text{var}[\mu] = \frac{ab}{(a+b)^2(a+b+1)}$$

```python
import matplotlib.pyplot as plt
import numpy as np
plt.style.use("seaborn")
```

## Probability Density Function 
$$\frac{x^{\alpha-1}(1 - x)^{\beta - 1}}{B(\alpha, \beta)}$$

where 
$$B(\alpha, \beta) = \frac{\Gamma(\alpha)\Gamma(\beta)}{\Gamma(\alpha + \beta)}$$

and $\Gamma$ is the `gamma function`.

```python
import scipy as sp
import scipy.special
import scipy.stats

pairs = [
    [0.5, 0.5],
    [5, 1],
    [1, 3],
    [2, 2],
    [2, 5]
]

x = np.linspace(0, 1, 100)
plt.figure(figsize=(16, 8))

for (a, b) in pairs:
    dist = sp.stats.distributions.beta(a, b)
    y = dist.pdf(x)
    plt.plot(x, y, label=f"a={a}, b={b}");
plt.legend()
plt.title("Probability density function");
```

## Cumulative Distribution Function (CDF)

```python
import scipy as sp
import scipy.special
import scipy.stats

pairs = [
    [0.5, 0.5],
    [5, 1],
    [1, 3],
    [2, 2],
    [2, 5]
]

x = np.linspace(0, 1, 100)
plt.figure(figsize=(16, 8))

for (a, b) in pairs:
    dist = sp.stats.distributions.beta(a, b)
    y = dist.cdf(x)
    plt.plot(x, y, label=f"a={a}, b={b}");
plt.legend()
plt.title("Cumulative distribution function");
```
