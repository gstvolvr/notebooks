# Law of Large Numbers
Theorem that describes the result of performing the same experiement, a large number of times. The `average` of the results obtained from a large number of trials should be close to the `expected value`. 

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
sns.set()


x = np.arange(1000)
s = []
for _ in x:
    r = np.random.rand()
    if r > 0.5:
        s.append(1)
    else:
        s.append(0)

plt.plot((pd.Series(s).cumsum() / x)[1:]);
```
