# Weibull Distribution
Continuous probability distribution commonly used in `survival analysis`. 

Some of it's special cases include:
* $k=1$ — [exponential distribution]()
* $k=2, \lambda = \sqrt(2)\sigma$ — `rayleigh distribution`

If `X` represent <i>time to failure</i>, the `weibull distribution` gives a distribution in which the failure rate is proportional to a power of time.

We use the following variabels:
* $k$ — the shape parameter
    * $k\le 1$  failure rate decreases over time
    * $k=1$  failure rate is constant over time
    * $k\ge$  failure rate increases over time
* $\lambda$ — the scale parameter


```python
import matplotlib.pyplot as plt
import numpy as np
plt.style.use("seaborn")
```

### Probability Density Function (PDF)
$$
f(x) =\begin{cases}
\frac{k}{\lambda} (\frac{x}{\lambda})^{k-1}e^{-(\frac{x}{\lambda})^k} & x \ge 0\\
0 & x \le 0\\
\end{cases}
$$

`scipy` has a `stats.exponweib` which isn't really what we want

```python
import scipy as sp
import scipy.special
import scipy.stats

def pdf(x,lamb,k):
    return (k / lamb) * (x / lamb)**(k-1) * np.exp(-(x/lamb)**k)

plt.figure(figsize=(16, 8))
x = np.linspace(0, 2.5, 100)
for k in [0.5, 1., 1.5, 5]:
#     dist = sp.stats.exponweib(a=1., c=1., scale=k)
    y = [pdf(i, 1., k) for i in x]
    plt.plot(x, y, label=f"k = {k}, lambda = 1")
plt.legend();
```
