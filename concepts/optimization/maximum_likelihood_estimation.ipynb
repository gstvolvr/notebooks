# [Maximum Likelihood Estimation (MLE)](https://brilliant.org/wiki/maximum-likelihood-estimation-mle/)
A technique used for estimating parameters of a given distribution, using observed data. For example, if we know that a process follows a `normal distribution`, but we don't know the `mean` or `variance`, we can use `MLE` to estimate them with a sample of the population. We do so by finding values for both parameters that make our observations most likely to have occured.

#### Example


If there are `N` planes, and we've randomly sampled 3, we can calculate the probability of that happening by:<br>
$$P(x_4, x_{17}, x_9) = 3! \frac{1}{N} \frac{1}{N-1} \frac{1}{N-2}$$

Since we're randomly picking from a population without replacement, and account for the order that we can pick them in (i.e. $3!$).

Looking at the possible answers, we can input each value and see which gives us the highest probability. In this case, the answer is `17`. There is also an intuitive explanation to this, which is that the chance of randomly picking any single plane decrease as `N` increases.
