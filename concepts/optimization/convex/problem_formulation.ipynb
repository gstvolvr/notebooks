# Problem Formulation

<ol>
    <li>Encode decisions/actions as `decision variables` whose value we are seeking</li>
    <li>Identify relevant `problem data`</li>
    <li>Express `constraints` on the values of the `decision variables` as mathematical relationships between variables and `problem data`</li>
    <li>Express the `objective function` in terms of the `decision variables` and the `problem data`</li>
</ol>
 

## Examples
<ol>
<li>Given a $1\times1$ cardboard, cut out corners and fold to make a box of maximum volume.

<b>decision:</b> How much to cut<br>
<b>alternatives:</b> $0 \le x \le \frac{1}{2}$<br>
<b>best - max volume:</b> $V(x) = x(1 - 2x)^2$<br>

$$max(1 - 2x)^2$$
$$s.t.~~~0 \le x \le \frac{1}{2}$$
</li>

## Generic Formulation
<ul>
    <li>Express all `decision variables` as one `decision vector`: $\mathbf{x} \in \mathbb{R}^n$</li>
    <li>The `objective function` as a vector argument: $f: \mathbb{R}^n \rightarrow \mathbb{R}$</li>
    <li>The constrains define the set of values the decision vector can take:<br> $$X = { x\in\mathbb{R}^{n-p}\times\mathbb{Z}^p: g_i(x)\le b_i~~i=1,\dots,m}$$
    </li>
</ul>


Here the $x \in \mathbb{R}^{n-p}\times\mathbb{Z}^p$ means that the first $n-p$ components are continuous variables, and the last $p$ are integers. 

$$min~f(x)~~s.t.~~x \in X$$

### Visualizing constraints
$g_1(x) = x^2 + y^2$<br>
$g_2(x) = x^3 - y$<br>
$X = { x: g_1(x) \le 4, g_2(x) \le 0 }$

```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
plt.style.use("seaborn")

plt.figure(figsize=(8, 8))

### plot unit circle
phi = np.linspace(0, 2*np.pi, 1000)
r = 2
x, y = r * np.cos(phi), r * np.sin(phi)
plt.plot(x, y)
plt.fill_between(x, y, alpha=0.2)

idx = [(i, j) for i, j in zip(x, y) if i + j <= 4]

### plot cubic function
def g2(x, y): return x**3 - y 
x = y = np.linspace(-10, 10, 1000)

Z = np.zeros((1000, 1000))
for i, j in enumerate(y):
    Z[i, :] = g2(x, j)

plt.plot(x, g2(x, 0));
plt.fill_between(x, g2(x, 0), g2(x, -100), alpha=0.2)

plt.grid(b=True, which='major', color='w', linewidth=1.0)
plt.xlim(-4, 4)
plt.ylim(-4, 4)
plt.axhline(y=0, color='k');
plt.axvline(x=0, color='k');
```

### Problem Classifications

|Functions|Variable Domains|Problem Type|Difficulty|
|-------:|-----|-----|-----|
|All linear|continuous variables|linear program|easy|
|Some nonlinear|continuous variables|nonlinear  program|$O(n~logn)$|
|Linear/nonlinear|some discrete|integer program or discrete optimization problem|hard|
