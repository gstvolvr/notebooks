# Optimality Conditions
---
#### TL;DR
The following are necessary but not sufficient
* gradient at $x^*$ as to be `0`
* `Hessian` is positive semi-definitive
---

## First Order Optimality Conditions
What are the conditions under with $x^*$ is a local optima?

We basically want to check when the following $x^*$ is a local optimal soultion. 

From Taylor's expansion:
$$f(x^*+\Delta x) = f(x^*) + \frac{\nabla f(x^*)^T \Delta x}{1!} + \frac{\nabla^2 f(x^*)^T \Delta x}{2!} + \dots$$

Where $\nabla^2 f(x^*)^T$ is the [hessian matrix](../linear-algebra/hessian_matrix.ipynb)

If $x^*$ is a local minima, then $f(x^* + \Delta x) \ge f(x^*)$ for $\Delta x$ succifiently small, i.e.

$$\frac{\nabla f(x^*)^T \Delta x}{1!} \ge 0$$

But $\Delta x$ can be positive or negative. Then we must have 
$$\nabla f(x^*) = 0$$

The above argument also holds for local maxima, i.e. if $x^*$ is a local maxima or minima, then $\nabla f(x^*) = 0$. 

A point where the gradient vanishes is called a stationary point. 

## Second Order Optimality Conditions
if $x^*$ is a local $min$, we have $\nabla f(x^*)^T \Delta x = 0$. Then for $f(x^* + \Delta x)\ge f(x^*)$ to hold for all $\Delta x$ we must have 
$$ \Delta x^T \nabla^2 f(x^*) \Delta x \ge 0$ for all $\Delta x$$

This is, $\nabla^2 f(x^*)$ is `positive semi-definitive`. 

### Example

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
def f(x): return x**3
def fp(x): return 3*x**2
def fpp(x): return 6*x

x = np.linspace(-10, 10, 1000)
plt.plot(x, f(x), label="function");
plt.plot(x, fp(x), label="1st order derivative");
plt.plot(x, fpp(x), label="2nd order derivative");
plt.legend();
```

We can see that point `(0, 0)` we meet both the optimality conditions, but we are at an `inflection point`, and not a `local optima`.
