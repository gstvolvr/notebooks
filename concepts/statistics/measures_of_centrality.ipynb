# Descriptive Statistics

How can we summarize numeric data? 

```python
import numpy as np
from scipy import stats
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()

data = np.random.gamma(50, size=1000)
fig = plt.figure(figsize=(12, 6))
plt.plot(data);
plt.title("random gamma distribution of values betten [0, 100]");
```

## Averages
There's different types of averages in statistics. Let's start with the most basic. 

---

### Arithmetic Mean
We add up all the values in our data, and divide by the number of data points. 

$$\frac{1}{n}\sum_{i=1}^n x_i$$

Where $n$ is the total number of data points, and $x_i$ is the $i^{th}$ value in our data.


```python
print(f"mean: {np.mean(data): 0.2f}")
```

### Median
We order our data, and choose the middle value. If $n$ is even, we take the `arithmetic mean` of the two middle numbers. This is particularly useful when we have data that contains outliers, as those outliers can [skew](skewness.ipynb) the data. 

```python
print(f"median: {np.median(data): 0.2f}")
```

### Mode
The most common number in a dataset. 

```python
print(f"mode: {stats.mode(data)[0][0]:0.2f}")
```
