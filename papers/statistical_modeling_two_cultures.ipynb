# [Statistical Modeling: The Two Cultures](https://projecteuclid.org/download/pdf_1/euclid.ss/1009213726)


There are two different academic “cultures” when it comes to prediction. They’re both trying to extract information about how input variables and response variables are related; and later use this information to make prediction on new input data. They just do it in different ways.

#### Data modeling culture (stats)
Makes the assumption that data are generated by a stochastic data model (e.g. assume that the data are normally distributed with ...)

Validation: 
Goodness of fit tests (e.g. Chi Square)
Residual tests


#### Algorithmic modeling culture (ML)
Doesn’t make any assumptions about the data model. It approximates a function from input x to output y using some form of optimization.


Validation:
Precision (aka PPV or specificity)
Ratio of positive predictions that are correct
Recall (aka sensitivity or true positive rate)
Ratio of positive instances that are correct
ROC (receiver operator curve) and AUC (area under the Curve)
Recall vs. False positive rate

It’s important to know that there’s a tradeoff between precision and recall; both can’t be maximized. This is depicted in the “precision and recall curve”. 


Often, ML models will be less interpretable than their statistical counterparts but can be more accurate.
